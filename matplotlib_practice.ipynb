```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, 'green', marker="D", linewidth="2")
ax.legend('Boston')
ax.set_title('PyBer Fare by Month')
ax.set_ylim(5, 50)
ax.grid()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Create the bar
plt.bar(x_axis, y_axis, color='green', label='Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the horizontal bar
plt.barh(x_axis, y_axis, color='magenta', label='Boston')
plt.gca().invert_yaxis()
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="k", label="New York")
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan")
ax.legend("Chicago")
ax.set_title("PyBer Fare by Month")
ax.set_ylabel("Fare ($)")
ax.set_xlabel("Date")
ax = plt.gca()
ax.set_ylim(ax.get_ylim()[::-1])
```

```python
# Create a scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create a scatter plot
plt.scatter(y_axis, x_axis, c="red")
plt.legend("Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare ($)")
plt.gca().invert_yaxis()
plt.xlim(0, 50)
```

```python
# Make the y axis larger
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

# Create a bubble chart
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Make the bubble chart
plt.scatter(x_axis, y_axis, s= [i * 3 for i in y_axis])
```

```python
# Create scatter plot in object-oriented interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a bubble chart in object-oriented interface
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], color="skyblue", alpha=0.2, 
           edgecolor="k")
ax.legend("Boston")
ax.set_title("PyBer Fares by Month")
ax.set_xlim(0, 50)
ax = plt.gca()
ax.set_ylim(ax.get_ylim()[::-1])
```

```python
colors = ['lightcoral','peachpuff','gold','palegoldenrod','greenyellow','mediumseagreen','paleturquoise','lightskyblue','slateblue','plum','orchid','palevioletred']
plt.subplots(figsize=(8,8))
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
explode = [0,0,0.3,0,0,0,0.3,0,0,0,0,0]
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(8)
ax.pie(y_axis, labels=x_axis, colors=colors, explode=explode, shadow=True, autopct='%1.1f%%', startangle=90, counterclock=False)
plt.show()
```
